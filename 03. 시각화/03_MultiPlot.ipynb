```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rcParams['axes.unicode_minus'] = False
plt.rc('font', family='NanumBarunGothic')
```

## Figure 객체

```python

np.random.seed(2021)
f1 = plt.figure(figsize=(10, 2))
plt.title("figure size : (10, 2)")
plt.plot(np.random.randn(100))
plt.grid()
plt.show()
```

## Axes 객체와 subplot

```python
x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)
```

```python

plt.subplot(2, 1, 1)      # Axes
plt.plot(x1, y1, 'yo-')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')
plt.grid()

plt.subplot(2, 1, 2)      # Axes
plt.plot(x2, y2, 'r.-')
plt.xlabel('time (s)')
plt.ylabel('Undamped')
plt.grid()

plt.tight_layout()
plt.show()
```

```python
np.random.seed(2021)

plt.subplot(221)            # 행의 갯수, 열의 갯수, 순서
plt.plot(np.random.rand(5))
plt.title("axes 1")
plt.grid()

plt.subplot(222)
plt.plot(np.random.rand(5))
plt.title("axes 2")
plt.grid()

plt.subplot(223)
plt.plot(np.random.rand(5))
plt.title("axes 3")
plt.grid()

plt.subplot(224)
plt.plot(np.random.rand(5))
plt.title("axes 4")
plt.grid()

plt.tight_layout()
plt.show()
```

- subplots()

```python

_, axes = plt.subplots(2, 2)    # 2 x 2 array

np.random.seed(2021)
axes[0, 0].plot(np.random.rand(5))
axes[0, 0].set_title("axes 1")
axes[0, 0].grid()
axes[0, 1].plot(np.random.rand(5))
axes[0, 1].set_title("axes 2")
axes[0, 1].grid()
axes[1, 0].plot(np.random.rand(5))
axes[1, 0].set_title("axes 3")
axes[1, 0].grid()
axes[1, 1].plot(np.random.rand(5))
axes[1, 1].set_title("axes 4")
axes[1, 1].grid()

plt.tight_layout()
plt.show()
```

```python
np.random.seed(2021)
fig, axes = plt.subplots(2,2, figsize=(6,4))    # 개별 그림의 사이즈

for i in range(4):
    ax = axes[i//2, i%2]    # 행 인덱스 i//2, 열 인덱스 i%2
    ax.plot(np.random.rand(5))
    ax.set_title(f'axes {i+1}')
    # ax.set_xlabel('X-축'); ax.set_ylabel('Y-축')
    ax.grid()

fig.suptitle('4개의 그림', fontsize=14)
plt.tight_layout()
plt.show()
```
