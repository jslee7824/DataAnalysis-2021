```python
import numpy as np
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rcParams['axes.unicode_minus'] = False
plt.rc('font', family='NanumBarunGothic')
```

```python
iris = sns.load_dataset("iris")    # 붓꽃 데이터
titanic = sns.load_dataset("titanic")    # 타이타닉호 데이터
tips = sns.load_dataset("tips")    # 팁 데이터
flights = sns.load_dataset("flights")    # 여객운송 데이터
```

## 1. 카운트플롯

```python
sns.countplot(x='class', data=titanic)
plt.title('타이타닉호의 각 클래스별 승객 수')
plt.show()
```

```python

sns.countplot(x="day", data=tips)
plt.title("요일별 팁을 준 횟수")
plt.show()
```

## 2. 다차원 실수형 데이터
- pairplot()

```python
sns.pairplot(iris)
plt.show()
```

```python
sns.pairplot(iris, hue='species', markers=['o','s','D']);
```

## 3. 2차원 카테고리 데이터

- heatmap()

```python
tt = titanic.pivot_table(index='class', columns='sex', aggfunc='size')
tt
```

```python

sns.heatmap(tt, annot=True, fmt='d')
plt.title('Titianic호의 선실별 남녀 승객수')
plt.show()
```

```python

sns.heatmap(tt, annot=True, fmt='d', cmap=sns.light_palette('purple'))
plt.title('Titianic호의 선실별 남녀 승객수')
plt.show()
```

```python
flights_passengers = flights.pivot("month", "year", "passengers")
plt.figure(figsize=(10,8))
plt.title("연도, 월 별 승객수에 대한 Heatmap")
sns.heatmap(flights_passengers, annot=True, fmt="d", linewidths=1)
plt.show()
```

## 4. 2차원 복합 데이터
- barplot()

```python
tips.head()
```

```python
sns.barplot(x='day', y='total_bill', data=tips)
plt.title("요일 별 전체 요금의 평균")
plt.show()
```

```python
sns.barplot(x='day', y='total_bill', hue='sex', data=tips)
plt.title("요일 별, 성별 전체 요금의 평균")
plt.show()
```

```python
sns.boxplot(x="day", y="total_bill", data=tips)
plt.title("요일 별 전체 팁의 Box Plot")
plt.show()
```

```python

sns.boxplot(x="day", y="total_bill", hue='sex', data=tips)
plt.title("요일 별/성별 전체 팁의 Box Plot")
plt.show()
```
